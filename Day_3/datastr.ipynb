```python
List = [1, 2, 3, 4, 5]
Tuple = (1, 2, 3, 4, 5)
Set = {1, 2, 3, 4, 5}   
```

```python
print(List)
print(Tuple)
print(Set)  
```

```python
# indexing zero based indexing
print(List[0])  
print(Tuple[1])  
print(Set)  #here indexing is not possible
print(List[-1])  #negative indexing -> last element

```

```python
# functions of List
List.append(7)
print(List)
List.insert(2, 6)  #index, value
print(List)
```

```python
List.remove(3) # it wiill remove first occurance of 3
print(List)
```

```python
List2 = List
print(List2)
List2.append(8)
print(List)
print(List2) # both sharing same memory location
```

```python
List3 = List.copy()  # creating a copy of List
List3.append(9)
print(List)
print(List3) #-> not effect in List(original list)
```

##  Slicing

```python
 # slicing works only in List and Tuple not in Set
print(List[1:4])  # from index 1 to index 3 and 4 is not included

print(Tuple[0:3])  


print(List[0:3:2])  
```

## Dictionaries

```python
dict = {"name":"abhijeet", "city":"ghaziabd", "hometown":"Kanpur"}
print(dict)
print(dict["city"])
dict["age"] = 25  # adding new key value pair
print(dict)
```

```python
#key is unque , value can be duplicate
dict2 = {"name":"rahul", "city":"ghaziabd", "hometown":"Kanpur","hometown":"Delhi"}
# takes the latest change for duplicate key
print(dict2)
```

```python
# length of data structures
print(len(dict))
print(len(List))
print(len(Tuple))
print(len(Set)) 
```

```python
print(type(dict))
```

```python
# finding the key in dictionary
print(dict.keys())
# finding the values in dictionary
print(dict.values())
```
