```python
# Data Structs (Lists, Dicts, Sets, Tuples, Slicing)
# Lists, Tuples, Sets, Dictionaries
```

```python
List = [1, 2, 3, 4, 5]
Tuple = (1, 2, 3, 4, 5)
Set = {1, 2, 3, 4, 5}   
```

```python
print(List)
print(Tuple)
print(Set)  
```

```python
# indexing zero based indexing
print(List[0])  
print(Tuple[1])  
print(Set)  #here indexing is not possible
print(List[-1])  #negative indexing -> last element

```

```python
# functions of List
List.append(7)
print(List)
List.insert(2, 6)  #index, value
print(List)
```

```python
List.remove(3) # it wiill remove first occurance of 3
print(List)
```

```python
List2 = List
print(List2)
List2.append(8)
print(List)
print(List2) # both sharing same memory location
```

```python
List3 = List.copy()  # creating a copy of List
List3.append(9)
print(List)
print(List3) #-> not effect in List(original list)
```

```python
List3.clear()
print(List3)    
```

```python
List2.count(2)  # counts how many times 2 is present in List2
print(List2)
List2.extend([10,13,18])
print(List2)

```

```python
Tuple[0]  # indexing in Tuple
print(Tuple)

Tuple.count(2)  # counts how many times 2 is present in Tuple
print(Tuple)  

Tuple.index(3) # how many times 3 is present in Tuple
print(Tuple)

Tuple2 = Tuple
print(Tuple2)

Tuple2 += (6,7)  # it will create a new Tuple
print(Tuple)

```

```python
Set.add(6)
print(Set)
```

```python
Set.add(6)
print(Set) # not include duplicate values

Set.remove(3)
print(Set)
```

##  Slicing

```python
 # slicing works only in List and Tuple not in Set
print(List[1:4])  # from index 1 to index 3 and 4 is not included

print(Tuple[0:3])  


print(List[:3:2])  
```
