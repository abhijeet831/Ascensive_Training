```python
class Dog:
    species = "Canine"

    def __init__(self,name,age):
        self.name = name 
        self.age = age

dog1 = Dog("Buddy",3)
print(dog1.name)
print(dog1.age)
print(dog1.species)
```

```python
class parent:
    def __init__(self):
        self.value = "inside Parent"

    def show(self):
        print(self.value)

class child(parent):
    def __init__(self):
        self.value = "inside child"

    def show(self):
        print(self.value)


obj1= parent()
obj2 =child()
obj1.show()
obj2.show()
```

```python
class vehicle:
    def __init__(self,name,model):
        self.name = name
        self.model = model

    def car(self):
        print(self.name , self.model)
    def bike(self):
        print(self.name , self.model)

car1 = vehicle("honda",2019)
car2 = vehicle("hyundai","2022")

bike1 = vehicle("yamaha",2022)
bike2 = vehicle("hero",2021)

car1.car()
car2.car()
# print("\n")
bike1.bike()
bike2.bike()
```

```python
# use of super()

class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)   # call parent constructor
        self.breed = breed

dog1 = Dog("Tommy", "bull dog")

print(dog1.name)
print(dog1.breed)
```

```python
class person:
    def __init__(self,fname,lname):
        self.firstname = fname
        self.lastname = lname

    def printname(self):
        print(self.firstname , self.lastname)

x = person("John","Doe")
x.printname()
```

```python
# encapulation example
class bank:
    def __init__(self):
        self.__balance = 0
    
    def deposit(self,amount):
        self.__balance += amount
    
    def show_balance(self):
        print(self.__balance)


b = bank()
b.deposit(10000)
b.show_balance()

# it can be accessed using name mangling
print(b._bank__balance) 
```
