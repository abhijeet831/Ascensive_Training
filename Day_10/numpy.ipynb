```python
import numpy as np

# array with index
arr = np.array([
    [-1,  2,  0,  4],
    [ 4, -5,  6,  0],
    [ 1,  2,  3,  4]
])

# first 2 rows and alternate columns
arr2 = arr[:2, ::2]
print("first 2 rows and alternate columns")
print(arr2)

# elements at specific indices
arr3 = arr[[1, 1, 0, 2], [1, 2, 3, 3]]
print("\nElements at indices (1,1), (1,2), (0,3), (2,3):", arr3)
```

```python
a = np.array([[1,2], [3,4], [5,6]])

b = np.array([[7,8], [9,10], [11,12]])

print("Adding 1 to every element:",a+1)

print("\nSubtracting 1 from every element:",a-1)
print("\nElement-wise addition of two arrays:",a+b)
print("\nElement-wise multiplication of two arrays:",a*b)
print("\nElement-wise division of two arrays:",a/b)

```

```python
x = np.array([1,2])
print(x.dtype)

x = np.array([1.0,2.0])
print(x.dtype)

x = np.array([1,2],dtype=np.int64)
print(x.dtype)
```

```python
# import numpy as np
print(np.arange(1,10))
print(np.linspace(1,10,5))
print(np.zeros((3,4)))
print(np.ones((2,3,4),dtype=np.int16))
print(np.random.rand(5))

print(np.random.randint(5,size=10))
```

```python
print(np.zeros([4,3],dtype=np.int32)) # 4-rows and 3-columns
print(np.ones([4,3],dtype = np.int32))

print(np.full([2,2],67,dtype=int))
print(np.eye(4))
```

```python
arr = np.asarray([1,2,3,4,5])
print(arr.size)
print(arr.shape)
print(len(arr))
print(arr.dtype)
print(arr.ndim)

# arr = arr.astype('')
```

```python
# change dtype to float
arr = arr.astype('float64')
print(arr.dtype)
print("Datatype:",arr.dtype)

lis = arr.tolist()
print(lis)
```

```python
a = np.array([1,4,9,16,25])

print(np.sqrt(a))
print(np.log(a))
print(np.exp(a))
```

```python
print(np.sin(a))
print(np.cos(a))
print(np.tan(a))
print(np.max(a))
print(np.mean(a))
```
