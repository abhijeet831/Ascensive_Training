### Def, Args/Kwargs, Lambda, Return types

```python
def evenodd(num):
    if num % 2 == 0:
        return "even"
    else:
        return "odd"

print(evenodd(10))
print(evenodd(5))
```

```python
# Default parameter
def myFun(x,y=50):
  print(x)
  print(y)
myFun(10)
```

```python
def myFun(x,y=50):
  print(x)
  print(y)
myFun(10,60)
```

```python
# keyword argument
def student(name,student_id):
  print(name,student_id)
student(name = "abhijeet",student_id=101)
student(name = "singh",student_id=102)
```

```python
#positional argument
def student(name,student_id):
  return(name,student_id)

```

```python
print(student("abhijeet",101))
print(student(101,"abhijeet"))
```

```python
def myfun(*args):
    return sum(args)
myfun(1,2,3,4)

```

```python
def myfun2(**kwargs):
    for key , value in kwargs.items():
        print(f"{key} == {value}")

myfun2(first = 'geeks', mid = 'for',last = 'geeks')
```

```python
def f1():
    s = "helloworld"
    print("f1")
    def f2():
        print(s)
        print("f2")
    
    f2()

f1()
```

```python
# lambda functions

cub_l = lambda x : x*x*x
cub_l(5)

```

```python
evenodd = lambda x: "even" if x % 2 == 0 else "odd"
evenodd(10)
```

```python
evenodd(11)
```

```python
nums = [1,2,3,4,5,6,7,8,9]
squares = list(map(lambda x: x*x ,nums))
print(squares)
```

```python
evenodd = list(filter(lambda x: x%2==0,nums))
print(evenodd)
```

```python
# recursive functions

def factorial(n):
    if n == 0:
        return 1
    else:
        # print(n)
        return n * factorial(n-1)
    
print(factorial(4))
```

```python
# pass by value 
#-> in pass by value the changes made to the parameter inside the function do not affect the argument.

def myfun3(x):
    x = 20
    print("x",x)

y=10
myfun3(y)
print("y",y)


```

```python
#pass by reference
# in pass by reference, the actual variable is passed to the function.
# So any changes made to the parameter inside the function will affect the original variable.

def myfun(x):
    x[0] = 20

lst = [10,20,30,40]
myfun(lst)
    
```

```python
def squares(nums):
    return [x**2 for x in nums] 
squares([1,2,3,4,5])
```

```python
def squares(nums):
    return nums**2 
squares(10)
```
